# Descriptive Statistics Before and After Winsorization

This notebook builds the paper and appendix descriptive-statistics tables from the prepared public and private nonprofit analysis panel. Winsorization is applied only inside the table builder. The parquet panel is not modified.

```python
from pathlib import Path
import sys

import pandas as pd
from IPython.display import display

REPO = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
SRC = REPO / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

from coa_finaid_subs.descstat_tables import build_descstat_tables

pd.set_option("display.max_columns", 120)
pd.set_option("display.width", 160)
```

```python
PANEL = REPO / "outputs" / "analysis_panel" / "public_private_nonprofit" / "analysis_panel_coa_headroom_2009_2023_public_private_nonprofit.parquet"
CONFIG = REPO / "config" / "descstat_variables.csv"
OUTPUT_DIR = REPO / "outputs" / "descriptive_tables"

paths = build_descstat_tables(
    input_panel=PANEL,
    output_dir=OUTPUT_DIR,
    config=CONFIG,
    scope_label="public_private_nonprofit",
)
paths
```

## Paper Table

The paper table keeps the main variables and reports raw and winsorized means and standard deviations side by side.
